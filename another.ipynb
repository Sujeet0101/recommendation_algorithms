```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("hosteldata.csv")
df.head(20)
```

```python
features = ["District", "Place/City", "HostelType", "SingleSeater", "DoubleSeater", "3Seater", "4Seater", "Wifi",
            "HotWater", "Parking", "Laundry", "CCTV", "Fan"]
all_features = []
for index, row in df.iterrows():
  hostel_features = []
  for feature in features:
    hostel_features.append(row[feature])
  hostel_features = [str(feature) for feature in hostel_features]
  all_features.append(" ".join(hostel_features))
```

```python
vectorizer = CountVectorizer()
feature_matrix = vectorizer.fit_transform(all_features)
```

```python
def get_recommendations(District, Place_City, HostelType, SingleSeater, DoubleSeater, ThreeSeater, FourSeater, Wifi,
                        HotWater, Parking, Laundry, CCTV, Fan):
    
    if SingleSeater == '':
        SingleSeater = 0
    else:
        SingleSeater = int(SingleSeater)

    if DoubleSeater == '':
        DoubleSeater = 0
    else:
        DoubleSeater = int(DoubleSeater)

    if ThreeSeater == '':
        ThreeSeater = 0
    else:
        ThreeSeater = int(ThreeSeater)

    if FourSeater == '':
        FourSeater = 0
    else:
        FourSeater = int(FourSeater)
  


  # Create a list of the user's input
    input_features = [District, Place_City, HostelType, SingleSeater, DoubleSeater, ThreeSeater, FourSeater, Wifi,
                        HotWater, Parking, Laundry, CCTV, " ".join(Fan)]
  # Use CountVectorizer to convert the input into a matrix of token counts
    input_matrix = vectorizer.transform([" ".join(input_features)])
  # Calculate the cosine similarity between the input and all hostels
    similarity = cosine_similarity(input_matrix, feature_matrix)
  # Get the indices of the most similar hostels
    most_similar = similarity.argsort()[:, ::-1]

  # Select only the hostels with prices greater than or equal to the user input price
    similar_hostels = df.iloc[most_similar[0][:5]]
    selected_hostels = similar_hostels[(similar_hostels['SingleSeater'] >= 'SingleSeater') &
                                      (similar_hostels['DoubleSeater'] >= 'DoubleSeater') &
                                      (similar_hostels['ThreeSeater'] >= 'ThreeSeater') &
                                      (similar_hostels['FourSeater'] >= 'FourSeater')]
  # Return the selected hostels
    return selected_hostels

```

```python
get_recommendations(District='Bhaktapur', Place_City='', HostelType='Boys',SingleSeater='', 
                    DoubleSeater='', ThreeSeater='', FourSeater='9000', Wifi=[1],
                        HotWater=[0], Parking=[1], Laundry=[1], CCTV=[1], Fan=[0])
```
